```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cost_arr=np.load('VMBQC_learnt_by_channel_8_qubits_fig_4.npy')
```

```python
ave_cost=[]
std=[]
a=[]
total_cost_arr=[]


num_models=12
iterations=100

for r in range(num_models):
    total_cost_arr.append(cost_arr[r*(iterations+1):(r+1)*(iterations+1)])
    
for j in range((iterations+1)):
    a=[]
    for i in range(num_models):
        a.append(total_cost_arr[i][j])
    std.append(np.std(a))
    ave_cost.append(np.average(a))

```

```python
fig, ax = plt.subplots()
x=np.linspace(0,len(ave_cost),len(ave_cost))
y=ave_cost
yerr=std
ax.semilogy(x, y, 'k', color='navy',label='$\widetilde{\mathcal{E}}_c, 8 qubits$')
ax.fill_between(x, np.array(y)+np.array(yerr),np.array(y)+(-1)*np.array(yerr),
    alpha=0.5, edgecolor='dodgerblue', facecolor='dodgerblue')

plt.title('Fig. 4, learning distributions of VMBQC models')
plt.legend()
plt.show()
```
